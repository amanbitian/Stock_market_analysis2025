```python
import time
import yfinance as yf
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# Read the locally downloaded CSV file
## Link to nifty 500 : https://www.nseindia.com/market-data/live-equity-market?symbol=NIFTY%20500
nifty500 = pd.read_csv("MW-NIFTY-500-05-Jan-2025.csv")

# Display the first few rows
display(nifty500.head())
nifty500.columns
```

```python
# Remove '\n' and blank spaces from column names
nifty500.columns = nifty500.columns.str.replace(r'\n|\s', '', regex=True)

# Check the updated column names
print(nifty500.columns)
```

```python
nifty500['SYMBOL']=nifty500['SYMBOL']+'.NS'
nifty500.head(2)
```

```python
print(nifty500.shape)

display(nifty500.sort_values(by='365D%CHNG03-Jan-2024', ascending= False).head(5))
display(nifty500.sort_values(by='365D%CHNG03-Jan-2024', ascending= True).head(5))
```

```python
### Drop NIFTY 500.NS from the df
nifty500 = nifty500[nifty500['SYMBOL'] != 'NIFTY 500.NS']
nifty500.shape
```

```python
### Creating a database for nifty 500 stocks
start_time = time.time()

# Load list of Nifty 500 stock tickers
nifty_500_tickers = list(nifty500['SYMBOL'])

# Create an empty list to hold the data for all stocks
stock_data_list = []

# Define the date range (1st Jan 2015 to 5th Jan 2025)
start_date = '2015-01-01'
end_date = '2025-01-05'

# Fetch data for each stock
for ticker in nifty_500_tickers:
    stock = yf.Ticker(ticker)
    data = stock.history(start=start_date, end=end_date)
    data['Stock Ticker'] = ticker
    stock_data_list.append(data)

# Concatenate all the data into a single DataFrame
df = pd.concat(stock_data_list)

# Reset index to have Stock Ticker as one of the columns
df = df.reset_index()

display(df.head(2))

# End time
end_time = time.time()

# Execution time in seconds
execution_time = end_time - start_time
print(f"Execution time: {execution_time:.2f} seconds")
```

```python
df.to_csv('Nifty500_data.csv')
```

```python

# Set MultiIndex with Stock Ticker and Date
df.set_index(['Stock Ticker', 'Date'], inplace=True)

# Drop columns that are not required
df = df[['Open', 'Close', 'Volume', 'Adjusted Price']]

# Rename 'Adj Close' to 'Adjusted Price'
df.rename(columns={'Adj Close': 'Adjusted Price'}, inplace=True)

# Display the first few rows of the DataFrame
print(df.head())
```
